```python
import pandas as pd
import matplotlib.pyplot as plt

df_bookings = pd.read_csv('datasets/fact_bookings.csv')
df_rooms = pd.read_csv('datasets/dim_rooms.csv')
print(df_bookings.shape)
df_bookings.head()
```

```python
df_bookings = df_bookings[df_bookings.no_guests>0]
df_bookings.shape
```

```python
df_bookings.revenue_generated.max(),df_bookings.revenue_generated.min()
```

```python
avg , std_dev = df_bookings.revenue_generated.mean(), df_bookings.revenue_generated.std()
upper_limit = avg + 3*std_dev  #we can remove the data which is 3 standard deviation away from mean
lower_limit = avg - 3*std_dev
print(upper_limit,lower_limit)
```

```python
df_bookings[df_bookings['revenue_generated']>upper_limit]
```

```python
df_bookings[df_bookings['revenue_generated']<=lower_limit]
```

```python
df_bookings = df_bookings[(df_bookings.revenue_generated<upper_limit) & (df_bookings.revenue_generated>lower_limit)]
print(df_bookings.shape)
df_bookings.head()
```

```python
df_bookings.revenue_realized.max(),df_bookings.revenue_realized.min()
```

```python
avg_rr , std_dev_rr = df_bookings.revenue_realized.mean(), df_bookings.revenue_realized.std()
upper_limit_RR = avg_rr + 3*std_dev_rr  #we can remove the data which is 3 standard deviation away from mean
lower_limit_RR = avg_rr - 3*std_dev_rr
print(upper_limit_RR,lower_limit_RR)
```

```python
df_bookings[df_bookings['revenue_realized']>upper_limit_RR]
```

```python
df_agg_bookings = pd.read_csv('datasets/fact_aggregated_bookings.csv')
print(df_agg_bookings.shape)
df_agg_bookings.head()
```

```python
df_agg_bookings.info()
```

```python
df_agg_bookings['Occupancy_percentage'] = df_agg_bookings['successful_bookings']/df_agg_bookings['capacity']
```

```python
df_agg_bookings.head()
```

```python
df_agg_bookings['Occupancy_percentage'] = df_agg_bookings['Occupancy_percentage'].apply(lambda x: round(x*100,2))
df_agg_bookings.head()
```

```python
df_agg_bookings.groupby('room_category')['Occupancy_percentage'].mean()
```

```python
df_rooms.head()
```

```python
df = pd.merge(df_agg_bookings, df_rooms, left_on='room_category', right_on='room_id', how='left').head()
df
```
