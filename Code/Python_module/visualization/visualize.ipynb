```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('linechart.xlsx')
df.head(10)
plt.figure(figsize=(10,4))
plt.plot(df['Quarter'], df['Fridge'],color='red',label='Fridge')
plt.plot(df['Quarter'], df['Dishwasher'],color='Orange',label='Dishwasher')
plt.plot(df['Quarter'], df['Washing Machine'],label='Washing Machine')
plt.title('Appliance Sales')
plt.xlabel('Quarter')
plt.ylabel('Revenue in USD')
plt.legend()
plt.show()
```

```python
total_sales = df[["Fridge","Dishwasher","Washing Machine"]].sum()
```

```python
total_sales.index
```

```python
plt.figure(figsize=(10,4))
plt.pie(total_sales,labels=total_sales.index,autopct='%1.1f%%',startangle=140,explode=(0.1,0,0),shadow=True)
plt.show()
```

```python
df.plot(kind="bar",x="Quarter",y=["Fridge","Dishwasher","Washing Machine"],figsize=(10,4),title="Appliance Sales")
plt.figure(figsize=(10,4))
plt.show()
```
