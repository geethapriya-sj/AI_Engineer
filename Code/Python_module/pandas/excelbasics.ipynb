```python
def standarize_currency(currency):
    if currency == "$$" or currency == "Dollars":
        return "USD"
    return currency
    
```

```python
df_finance = pd.read_excel("movies_db.xlsx","financials",converters={"currency":standarize_currency})
df_finance.head(10)
```

```python
df_finance.currency.unique()
```

```python
df_merge = pd.merge(df_movies,df_finance,on="movie_id")
df_merge.head(10)
df_merge.to_excel("merged_movies.xlsx",sheet_name="movies_financials",index=False)
```
