```python
import csv

def calculate_rating_statistics(ratings):
    ratings = [float(r) for r in ratings if r]
    if not ratings:
        return None, None
    avg_rating = sum(ratings) / len(ratings)
    min_rating = min(ratings)
    max_rating = max(ratings)
    return avg_rating, min_rating, max_rating

with open("movies.csv","r")as f:
    contents = list(csv.reader(f))
    header = contents[0]
    data = contents[1:]
    print(header)
    print(data[:5])
```
